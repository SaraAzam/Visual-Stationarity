```
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/airline-passengers.csv')
X=df['Passengers']
#print(X)
Split=round((len(X)/2))
#print(Split)
X1,X2=X[0:Split],X[Split:]
#print(X1)
mean1=X1.mean()
mean2=X2.mean()
var1=X1.var()
var2=X2.var()
print('m1',mean1)
print('m2',mean2)
print('v1',var1)
print('v2',var2)
plt.hist(X)
plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df=pd.read_csv('/content/airline-passengers.csv')
X=df['Passengers']
lx=np.log(df['Passengers'])
print(lx)
Split=round((len(lx)/2))
#print(Split)
lX1,lX2=lx[0:Split],lx[Split:]
#print(X1)
mean1=lX1.mean()
mean2=lX2.mean()
var1=lX1.var()
var2=lX2.var()
print('m1',mean1)
print('m2',mean2)
print('v1',var1)
print('v2',var2)
plt.hist(lx)
plt.show()
```
